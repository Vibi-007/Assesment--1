```python
'''
Matrimonial application for happy life
Author: Vibhish R
Contributors: Vibhish R
Created on: 14-09-2022
Modified on: 14-09-2022
'''
import csv
import logging
data = ['uni_id','name', 'gender', 'age', 'occupation', 'hobby', 'likes', 'dislikes']
filename = "mat.csv"
def insertion():
    global data
    global filename
    l = []
    for x in data:
        value = input("Enter " + x + ": ")
        l.append(value)
    with open(filename, "a", newline = "") as file:
        writer = csv.writer(file)
        writer.writerows([l])
    print("Data saved successfully")

def modification():
    global data
    global filename
    uni_id = int(input("Enter uni_id to be update: "))
    n_user = None
    l2 = []
    with open(filename, "r") as file:
        reader = csv.reader(file)
        c = 0
        for row in reader:
            if len(row) > 0:
                n_user = c
                print("user at", n_user)
                l = []
                for x in data:
                    value = input("Enter " + x + ": ")
                    l.append(value)
                l2.append(l)
            else:
                l2.append(row)
            c += 1

def searching():
    global data
    global filename
    uni_id = input("Enter uni_id to search: ")
    with open(filename, "r") as file:
        reader = csv.reader(file)
        for row in reader:
            if len(row) > 0:
                if uni_id == uni_id[0]:
                    print("uni_id: ", row[0])
                    print("name: ", row[1])
                    print("gender: ", row[2])
                    print("age: ", row[3])
                    print("occupation: ", row[4])
                    print("hobby: ", row[5])
                    print("likes: ", row[6])
                    print("dislikes: ", row[7])
                    break
        else:
            print("uni_id not found")

def deletion():
    global data
    global filename
    uni_id = input("Enter uni_id to delete: ")
    a = False
    l2 = []
    with open(filename, "r") as file:
        reader = csv.reader(file)
        c = 0
        for row in reader:
            if len(row) > 0:
                if uni_id != row[0]:
                    l2.append(row)
                    c +=1
                else:
                    a = True
    if a is True:
        with open(filename, "w") as file:
            writer = csv.writer(file)
            writer.writerows(l2)
        print("uni_id", uni_id, "deleted successfully")
    else:
        print("not found")        
def menu():
    print("*" * 48)
    print("Welcome to Matrimonial system")
    print("*" * 48)
    print("1.Insertion")
    print("2.Modification")
    print("3.Searching")
    print("4.deletion")

while True:
    menu()
    try:
        choice = eval(input("Enter the  choice: "))
    except:
        logging.warning("enter the choice as number")
    if choice == 0:
        break
    elif choice == 1:
        insertion()        
    elif choice == 2:
        modification()       
    elif choice == 3:
          searching()           
    elif choice == 4:
          deletion()                        
    else:
        break
            

```
